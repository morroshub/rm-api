```python
# Obtener la lista de episodios desde la API
episodes_url = 'https://rickandmortyapi.com/api/episode'
response = requests.get(episodes_url)
episodes_data = response.json()

# Crear una lista de todas las URL de episodios
episode_urls = [episode['url'] for episode in episodes_data['results']]

characters_count = {}

# Iterar sobre cada URL de episodio
for episode_url in episode_urls:
    # Obtener el episodio
    response = requests.get(episode_url)
    episode_data = response.json()
    
    # Obtener los personajes del episodio
    characters = episode_data['characters']
    
    # Iterar sobre cada personaje del episodio
    for character in characters:
        # Realizar solicitud a la API para obtener información del personaje
        response = requests.get(character)
        character_data = response.json()
        character_name = character_data['name']
        
        # Incrementar el contador de apariciones del personaje
        if character_name in characters_count:
            characters_count[character_name] += 1
        else:
            characters_count[character_name] = 1

# Crear un DataFrame a partir del diccionario de recuento de personajes
df = pd.DataFrame.from_dict(characters_count, orient='index', columns=['Apariciones'])
df.index.name = 'Personaje'

# Ordenar los personajes según su contador de apariciones en orden descendente
top_15_characters = df.sort_values('Apariciones', ascending=False).head(15)

# Imprimir los personajes
print("Los 15 personajes que más aparecen en los episodios son:")
top_15_characters
```

```python
#Hagamos una lista diferenciando a los personajes entre humanos y otras especies :

url = 'https://rickandmortyapi.com/api/episode/1'
r = requests.get(url)
j = r.json()
personajes = j['characters']

list_names_human = []
list_names_otherspecies = []

for personaje in personajes:
    req = requests.get(personaje)
    js = req.json()
    name = js['name']
    if js['species'] == 'Human':
        list_names_human.append(name)
    else:
        list_names_otherspecies.append(name)
        
print("Personajes Humanos:")
print(list_names_human)
print("Personajes de Otras Especies:")
print(list_names_otherspecies)
```

```python
### Hagamos un dataframe con PANDAS! 

# Crear un DataFrame a partir de las listas
df_human = pd.DataFrame({'Name': list_names_human, 'Species': 'Human'})
df_otherspecies = pd.DataFrame({'Name': list_names_otherspecies, 'Species': 'Other Species'})

# Concatenar los DataFrames en uno solo
df = pd.concat([df_human, df_otherspecies], ignore_index=True)


df
```
